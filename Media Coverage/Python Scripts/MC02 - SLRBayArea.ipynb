```python
# Set up your API key and import needed things
import os, mediacloud.api
from importlib.metadata import version
from dotenv import load_dotenv
import datetime as dt
from IPython.display import JSON
import pandas as pd
import bokeh.io
bokeh.io.reset_output()
bokeh.io.output_notebook()
MC_API_KEY = '3775cdee0e7968c994be1c001b3434ecddbdaa26' #may need to get a new API_Key from media cloud if we have exhausted this one
search_api = mediacloud.api.SearchApi(MC_API_KEY)
f'Using Media Cloud python client v{version("mediacloud")}'
```

```python
my_query = '("sea level rise" OR "sea-level rise" OR "SLR" OR "Aumento del nivel del mar") AND ("San Francisco" OR "SF Bay" OR "San Pablo Bay" OR "S.F. Bay" OR "Suisun Bay")'
start_date = dt.date(2023,1,1)
end_date = dt.date(2025,1,10) #Change date to the current date!!
CALIFORNIA_COLLECTION = 38380550 #this is the California Collection #, it is pulling data from local and national media outlets that cover California
results = search_api.story_list(my_query, start_date, end_date, collection_ids=[CALIFORNIA_COLLECTION])
```

```python
# fetch all the stories matching our query
all_stories = []
more_stories = True
pagination_token = None
while more_stories:
    page, pagination_token = search_api.story_list(my_query, dt.date(2023,1,1), dt.date(2025,1,10),#Change date to the current date!!
                                                   source_ids=[CALIFORNIA_COLLECTION],
                                                   pagination_token=pagination_token)
    all_stories += page
    more_stories = pagination_token is not None
len(all_stories)
```

```python
import csv
fieldnames = ['id', 'publish_date', 'title', 'url', 'language', 'media_name', 'media_url', 'indexed_date']
with open('story-list.csv', 'w', newline='',encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames, extrasaction='ignore')
    writer.writeheader()
    for s in all_stories:
        writer.writerow(s)
```
